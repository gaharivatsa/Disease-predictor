```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("D:\machine_learning\disease predictor\disease_symptom_dataset.xlsx")
df.head(35)
```

```python
df.drop('Count of Disease Occurrence',axis=1,inplace=True)
```

```python
for i,j in enumerate(df['Disease']):
    if not pd.isna(j):
        var = j
    if pd.isna(j):
        df['Disease'].iloc[i] = var

        
        
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.isna().any()
```

```python
df['Disease'] = df['Disease'].apply(lambda x:str(x.split('_')[-1]))
```

```python
for i in df['Symptom']:
    if type(i) != str:
        print(i,type(i))
```

```python
df['Symptom'] = df['Symptom'].apply(lambda x:str(x.split('_')[-1]))
```

```python
df.head(10)
```

```python
grouped_data = df.groupby('Disease')['Symptom'].agg(list).reset_index()
```

```python
grouped_data
```

```python
encoded_symptoms = pd.get_dummies(df['Symptom'])
```

```python
encoded_symptoms
```

```python
df_encoded = pd.concat([df['Disease'], encoded_symptoms], axis=1)
```

```python
df_encoded
```

```python
df_encoded.loc[df_encoded['pain chest']]
```

```python
df_encoded_grouped = df_encoded.groupby('Disease').max().astype(bool)

df_encoded_grouped = df_encoded_grouped.iloc[:,1:]
df_encoded_grouped.reset_index(inplace=True)
```

```python
df_encoded_grouped
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
df_encoded_grouped['Disease'] = label_encoder.fit_transform(df_encoded_grouped['Disease'])
```

```python
df_encoded_grouped
```

```python
X = df_encoded_grouped.drop(columns=['Disease'])
y = df_encoded_grouped['Disease']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.svm import SVC

svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)


y_pred = svm_model.predict(X_test)

```

```python
new_symptoms =  ['dysarthria', 'asthenia', 'speech slurred']


new_symptoms_df = pd.DataFrame(0, index=range(len(new_symptoms)), columns=X.columns)


for symptom in new_symptoms:
    new_symptoms_df[symptom] = 1


predicted_disease = svm_model.predict(new_symptoms_df)
predicted_disease_name = label_encoder.inverse_transform(predicted_disease)[0]
print("Predicted Disease:", predicted_disease_name)
```
